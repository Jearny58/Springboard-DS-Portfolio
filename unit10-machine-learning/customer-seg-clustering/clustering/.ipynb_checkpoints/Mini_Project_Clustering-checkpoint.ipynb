## K-Means Clustering

Recall that in K-Means Clustering we want to *maximize* the distance between centroids and *minimize* the distance between data points and the respective centroid for the cluster they are in. True evaluation for unsupervised learning would require labeled data; however, we can use a variety of intuitive metrics to try to pick the number of clusters K. We will introduce two methods: the Elbow method, the Silhouette method and the gap statistic.

### Choosing K: The Elbow Sum-of-Squares Method

The first method looks at the sum-of-squares error in each cluster against $K$. We compute the distance from each data point to the center of the cluster (centroid) to which the data point was assigned. 

$$SS = \sum_k \sum_{x_i \in C_k} \sum_{x_j \in C_k} \left( x_i - x_j \right)^2 = \sum_k \sum_{x_i \in C_k} \left( x_i - \mu_k \right)^2$$

where $x_i$ is a point, $C_k$ represents cluster $k$ and $\mu_k$ is the centroid for cluster $k$. We can plot SS vs. $K$ and choose the *elbow point* in the plot as the best value for $K$. The elbow point is the point at which the plot starts descending much more slowly.

```python
# your turn
# create a NumPy matrix 'x_cols' with only the columns representing the offers
x_cols = customer_table.iloc[:, 1:]
```

```python
# import KMeans from sklearn
from sklearn.cluster import KMeans
```

```python
k = range(2, 11)

# create list for values or inertia, or within-cluster sum-of-squares
inertia = []

for ks in k:
    # create a KMeans instance with ks clusters
    model = KMeans(n_clusters = ks, random_state = 1)
    
    # fit model to x_cols
    model.fit_predict(x_cols)
    
    # append the inertia to the list of inertia
    inertia.append(model.inertia_)
```

```python
# construct a plot showing SS for each K and pick K using this plot
plt.figure(figsize = (10, 8))
plt.plot(k, inertia)
plt.xlabel('Number of Clusters, K', fontsize = 'large')
plt.ylabel('Inertia, i.e. Within-Cluster Sum-of-Squares', fontsize = 'large')
plt.title('KMeans Inertia Graph')
plt.xticks(k);
```

Based on the graph above, the number of cluster I'm going to pick is 4 since after this point the benefits of adding more cluster is marginally less.

```python
# Make a bar chart showing the number of points in each cluster for k-means under the best  𝐾
model_4 = KMeans(n_clusters = 4)
customer_table['cluster'] = model_4.fit_predict(x_cols)
cluster_4_value_counts = customer_table.cluster.value_counts()
```

```python
plt.figure(figsize = (10, 8))
cluster_4_value_counts.plot(kind = 'bar')
plt.xlabel('Cluster #', fontsize = 'medium')
plt.xticks(rotation=0)
plt.ylabel('Total Number of Points', fontsize = 'medium')
plt.title('Number of Points in Each Cluster');
```

### Choosing K: The Silhouette Method

There exists another method that measures how well each datapoint $x_i$ "fits" its assigned cluster *and also* how poorly it fits into other clusters. This is a different way of looking at the same objective. Denote $a_{x_i}$ as the *average* distance from $x_i$ to all other points within its own cluster $k$. The lower the value, the better. On the other hand $b_{x_i}$ is the minimum average distance from $x_i$ to points in a different cluster, minimized over clusters. That is, compute separately for each cluster the average distance from $x_i$ to the points within that cluster, and then take the minimum. The silhouette $s(x_i)$ is defined as

$$s(x_i) = \frac{b_{x_i} - a_{x_i}}{\max{\left( a_{x_i}, b_{x_i}\right)}}$$

The silhouette score is computed on *every datapoint in every cluster*. The silhouette score ranges from -1 (a poor clustering) to +1 (a very dense clustering) with 0 denoting the situation where clusters overlap. Some criteria for the silhouette coefficient is provided in the table below.

<pre>

| Range       | Interpretation                                |
|-------------|-----------------------------------------------|
| 0.71 - 1.0  | A strong structure has been found.            |
| 0.51 - 0.7  | A reasonable structure has been found.        |
| 0.26 - 0.5  | The structure is weak and could be artificial.|
| < 0.25      | No substantial structure has been found.      |

</pre>
Source: http://www.stat.berkeley.edu/~spector/s133/Clus.html

Fortunately, scikit-learn provides a function to compute this for us (phew!) called [`sklearn.metrics.silhouette_score`](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_score.html). Take a look at [this article](http://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html) on picking $K$ in scikit-learn, as it will help you in the next exercise set.

```python
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm
import numpy as np

range_n_clusters = range(2, 11)
silhouette = []

for k in range_n_clusters:
    
    # estimate KMeans clustering and compute the silhouette scores
    kmeans = KMeans(n_clusters = k, random_state = 1)
    cluster_labels = kmeans.fit_predict(x_cols)
    silhouette_avg = silhouette_score(x_cols, cluster_labels, metric='euclidean')
    sample_silhouette_values = silhouette_samples(x_cols, cluster_labels)
    silhouette.append(silhouette_score)
    
    # Create a subplot with 1 row and 1 columns
    fig, ax1 = plt.subplots(1, 1)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(customer_table) + (k + 1) * 10])

    y_lower = 10
    
    for i in range(k):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        cluster_i = sample_silhouette_values[cluster_labels == i]
        size_cluster_i = cluster_i.shape[0]
        cluster_i.sort()
        
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / k)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, cluster_i, facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % k),
                 fontsize=14, fontweight='bold')

plt.show()
```

### Choosing $K$: The Gap Statistic

There is one last method worth covering for picking $K$, the so-called Gap statistic. The computation for the gap statistic builds on the sum-of-squares established in the Elbow method discussion, and compares it to the sum-of-squares of a "null distribution," that is, a random set of points with no clustering. The estimate for the optimal number of clusters $K$ is the value for which $\log{SS}$ falls the farthest below that of the reference distribution:

$$G_k = E_n^*\{\log SS_k\} - \log SS_k$$

In other words a good clustering yields a much larger difference between the reference distribution and the clustered data. The reference distribution is a Monte Carlo (randomization) procedure that constructs $B$ random distributions of points within the bounding box (limits) of the original data and then applies K-means to this synthetic distribution of data points.. $E_n^*\{\log SS_k\}$ is just the average $SS_k$ over all $B$ replicates. We then compute the standard deviation $\sigma_{SS}$ of the values of $SS_k$ computed from the $B$ replicates of the reference distribution and compute

$$s_k = \sqrt{1+1/B}\sigma_{SS}$$

Finally, we choose $K=k$ such that $G_k \geq G_{k+1} - s_{k+1}$.

### Aside: Choosing $K$ when we Have Labels

Unsupervised learning expects that we do not have the labels. In some situations, we may wish to cluster data that is labeled. Computing the optimal number of clusters is much easier if we have access to labels. There are several methods available. We will not go into the math or details since it is rare to have access to the labels, but we provide the names and references of these measures.

* Adjusted Rand Index
* Mutual Information
* V-Measure
* Fowlkes–Mallows index

See [this article](http://scikit-learn.org/stable/modules/clustering.html) for more information about these metrics.

## Visualizing Clusters using PCA

How do we visualize clusters? If we only had two features, we could likely plot the data as is. But we have 100 data points each containing 32 features (dimensions). Principal Component Analysis (PCA) will help us reduce the dimensionality of our data from 32 to something lower. For a visualization on the coordinate plane, we will use 2 dimensions. In this exercise, we're going to use it to transform our multi-dimensional dataset into a 2 dimensional dataset.

This is only one use of PCA for dimension reduction. We can also use PCA when we want to perform regression but we have a set of highly correlated variables. PCA untangles these correlations into a smaller number of features/predictors all of which are orthogonal (not correlated). PCA is also used to reduce a large set of variables into a much smaller one.

```python
#your turn; Usa PCA to plot clusters
# import PCA
from sklearn.decomposition import PCA

# create PCA instance
pca = PCA(n_components = 2, random_state = 1)

# apply fit_transform method of model to x_cols
pca_features = pca.fit_transform(x_cols)

# print shape of pca_features
pca_features.shape
```

```python
# assign column 0 of pca_features to xs
xs = pca_features[:, 0]

# assign column 1 of pca_features to ys
ys = pca_features[:, 1]
```

```python
# add pca_features to customer_table dateframe
customer_table['x'] = xs
customer_table['y'] = ys
```

```python
# create scatter plot of two principal components
plt.figure(figsize = (12, 8))
plt.grid(b = False)
plt.scatter(customer_table.x, customer_table.y, alpha = 0.8, c = customer_table.cluster, cmap = 'viridis')
plt.title('Scatter plot of Clustered Data')
plt.xlabel('Feature space - 1st Feature', size = 'x-large')
plt.ylabel('Feature space - 2nd Feature', size = 'x-large');
```

What we've done is we've taken those columns of 0/1 indicator variables, and we've transformed them into a 2-D dataset. We took one column and arbitrarily called it `x` and then called the other `y`. Now we can throw each point into a scatterplot. We color coded each point based on it's cluster so it's easier to see them.

## Other Clustering Algorithms

k-means is only one of a ton of clustering algorithms. Below is a brief description of several clustering algorithms, and the table provides references to the other clustering algorithms in scikit-learn. 

* **Affinity Propagation** does not require the number of clusters $K$ to be known in advance! AP uses a "message passing" paradigm to cluster points based on their similarity. 

* **Spectral Clustering** uses the eigenvalues of a similarity matrix to reduce the dimensionality of the data before clustering in a lower dimensional space. This is tangentially similar to what we did to visualize k-means clusters using PCA. The number of clusters must be known a priori.

* **Ward's Method** applies to hierarchical clustering. Hierarchical clustering algorithms take a set of data and successively divide the observations into more and more clusters at each layer of the hierarchy. Ward's method is used to determine when two clusters in the hierarchy should be combined into one. It is basically an extension of hierarchical clustering. Hierarchical clustering is *divisive*, that is, all observations are part of the same cluster at first, and at each successive iteration, the clusters are made smaller and smaller. With hierarchical clustering, a hierarchy is constructed, and there is not really the concept of "number of clusters." The number of clusters simply determines how low or how high in the hierarchy we reference and can be determined empirically or by looking at the [dendogram](https://docs.scipy.org/doc/scipy-0.18.1/reference/generated/scipy.cluster.hierarchy.dendrogram.html).

* **Agglomerative Clustering** is similar to hierarchical clustering but but is not divisive, it is *agglomerative*. That is, every observation is placed into its own cluster and at each iteration or level or the hierarchy, observations are merged into fewer and fewer clusters until convergence. Similar to hierarchical clustering, the constructed hierarchy contains all possible numbers of clusters and it is up to the analyst to pick the number by reviewing statistics or the dendogram.

* **DBSCAN** is based on point density rather than distance. It groups together points with many nearby neighbors. DBSCAN is one of the most cited algorithms in the literature. It does not require knowing the number of clusters a priori, but does require specifying the neighborhood size.

### Affinity Propagation
Often when we talk of clustering, k is not known which means that K-means clustering can be a little tricky. However, affinity propagation is a clustering technique that doesn’t require K and even defines K for you! 

With this technique, the data points can be seen as a network where all the points essentially communicate with all the other points. This is in an effort to find the exemplars, which are actual data points that best explain the other data points around them (i.e. are in the same cluster). A key point to remember is that each cluster has only one exemplar. 

Now the math gets a little complicated but there are two points to remember in this calculation:
- The first is the ‘responsibility’ matrix. This matrix reflects how well-suited one point is to be an exemplar for another point. 
- The second is the ‘availability’ matrix, which reflects how appropriate it would be for one point to choose another point as its exemplar. 

The inputs for this algorithm are:
- The number of iterations (convergence is reached i.e. the exemplars don’t change)
- The damping factor (the extent to which the current value is maintained relative to incoming values, helps to avoid numerical oscillations when updating these values)
- The preference, which indicates how strongly a data point thinks itself to be an exemplar. 


```python
# Your turn: Affinity propogation
from sklearn.cluster import AffinityPropagation
```

```python
affinityprop = AffinityPropagation(damping = 0.9, preference = -15)
affinity_labels = affinityprop.fit_predict(x_cols)
```

```python
plt.figure(figsize = (12, 8))
plt.grid(b = False)
plt.scatter(customer_table.x, customer_table.y, alpha = 0.8, c = affinity_labels, cmap = 'viridis')
plt.title('Scatter plot of Clustered Data')
plt.xlabel('Feature space - 1st Feature', size = 'x-large')
plt.ylabel('Feature space - 2nd Feature', size = 'x-large');
```

#### Spectral Clustering

The basic idea behind spectral clustering is that you define the similarity between any two data points however you want and then put them in a matrix. The cool thing is that you can define similarity in any way you want, which gives you more flexibility when compared to K-means. Another benefit is that it can theoretically work well for any group and not just data that is elliptically shaped. However, this flexibility can be a burden in that there are numerous ways to group points.

The math gets a little complicated and is beyond my current ability, but I’ll best try to explain it with the following steps:
- Define ‘similarity’ -- popular methods include Euclidean distance, a kernel function of Euclidean distance or a k-nearest neighbors approach
- Once you have a similarity matrix, create a Laplacian matrix (i.e. a matrix representation of a graph)
- Calculate the eigenvectors and eigenvalues of the Laplacian
- Use k-means algorithm on eigenvalues corresponding to the k-smallest eigenvectors
- The result will be k-clusters

While the math can be complicated, essentially it attempts to find a lower dimensional space that a standard clustering algorithm (such as K-means) can then be run on. It still has some of K-means weaknesses though, namely:
- We are still partitioning and polluting the clusters with noise
- We have to know the correct number of clusters or have to find some way to pick the ‘right one’

```python
# Your turn: Spectral clustering
```

```python
from sklearn.cluster import SpectralClustering
spectral_clustering = SpectralClustering(n_clusters = 4, random_state = 1)
spectral_labels = spectral_clustering.fit_predict(x_cols)
```

```python
plt.figure(figsize = (12, 8))
plt.grid(b = False)
plt.scatter(customer_table.x, customer_table.y, alpha = 0.8, c = spectral_labels, cmap = 'viridis')
plt.title('Scatter plot of Clustered Data')
plt.xlabel('Feature space - 1st Feature', size = 'x-large')
plt.ylabel('Feature space - 2nd Feature', size = 'x-large');
```

### Agglomerative Clustering

Agglomerative clustering is a bottom-up hierarchical clustering technique. What this means is that the algorithm starts by treating each object (i.e. data point) as a singleton cluster. Based on similarity, pairs of clusters are then merged. This process is iterated until all points are member of one single cluster that contains all objects.
What you get is a hierarchy, or a binary tree, of clusters branching down to the last layer which has a leaf for each point of the dataset. 

The linkage function is a key concept in making this hierarchy. What it does is take the distance information and then groups pairs of objects into clusters based on their similarity. The most basic version of this, single linkage, chooses the closest cluster to merge. There are more complect variations though, namely:
- Complete linkage: distance between clusters defined as the maximum value of all pairwise distances between the elements in cluster 1 and elements in cluster 2 (tends to produce compact clusters)
- Average linkage: distance between two clusters defined as the average distance between the elements in cluster 1 and elements in cluster 2
- Centroid linkage: distance between two clusters defined as the distance between the centroid for cluster 1 and the centroid for cluster 2
- Ward’s minimum variance method: minimizes total within-cluster variance; at each step, the pair of clusters with minimum between-cluster distance are merged

This technique has an advantage in that it does not presume the clusters to be globular. However it does have some weaknesses:
- Still have to choose the number of clusters which may not be obvious
- Still assuming all the data belongs in clusters

### DBSCAN

Known as ‘Density-based Spatial Clustering of Applications with Noise’, DBSCAN groups together points that are close to each other based on a distance measurement (usually Euclidean distance) and a minimum number of points. Essentially what it does is extract the ‘dense’ clusters and leaves the sparse background classified as ‘noise’. 

It has numerous advantages, including:
- Does not require one to specify the number of clusters a priori
- Can find arbitrarily shaped clusters 
- Has a notion of noise and is robust to outliers

But it also has some disadvantages:
- The quality depends on the distance measure, which can be rendered useless due to high-dimensional data
- Cannot cluster data well with large differences in densities
- If data and scale are not well understood, choosing meaningful distance threshold ε can be difficult.

Those two last points are evident in this specific instance. There is varying density and it was particularly hard to figure out what the ε should be. 

```python
# Your turn: DBSCAN
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(min_samples=3, eps = 1)
dbscan_labels = dbscan.fit_predict(x_cols)
```

```python
plt.figure(figsize = (12, 8))
plt.grid(b = False)
plt.scatter(customer_table.x, customer_table.y, alpha = 0.8, c = dbscan_labels, cmap = 'viridis')
plt.title('Scatter plot of Clustered Data')
plt.xlabel('Feature space - 1st Feature', size = 'x-large')
plt.ylabel('Feature space - 2nd Feature', size = 'x-large');
```
