## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

### 3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
import pandas as pd

q3_df = pd.read_json('data/world_bank_projects.json')
q3_df.head(3)
```

```python
import numpy as np
a1_fix = a1.replace('', np.nan)
a1_fix.isnull().sum()
```

```python
type(a1_fix.iloc[1,1])
```

```python
mask_11 = (a1_fix.name.isnull()) & (a1_fix.code == '11')
a1_fix.loc[mask_11, 'name'] = a1_fix.loc[mask_11, 'name'].apply(lambda x: 'Environment and natural resources management')
a1_fix.head(2)
```

```python
a1_fix.isnull().sum()
```

```python
def code_filler(df, code_num, code_str):
    mask = (df.name.isnull()) & (df.code == code_num)
    df.loc[mask, 'name'] = df.loc[mask, 'name'].apply(lambda x: code_str)
    return df
```

```python
a10 = code_filler(a1_fix, '10', 'Rural development')
a10[a10.code == '10'].isnull().sum()
```

```python
a10.isnull().sum()
```

```python
a8 = code_filler(a10, '8', 'Human development')
a8[a8.code == '8'].isnull().sum()
```

```python
a2 = code_filler(a8, '2', 'Public sector governance')
a2[a2.code == '2'].isnull().sum()
```

```python
a2.name.isnull().value_counts()
```

```python
a6 = code_filler(a2, '6', 'Social protection and risk management')
a6[a6.code == '6'].isnull().sum()
```

```python
a4 = code_filler(a6, '4', 'Financial and private sector development')
a4[a4.code == '4'].isnull().sum()
```

```python
a7 = code_filler(a4, '7', 'Social dev/gender/inclusion')
a7[a7.code == '7'].isnull().sum()
```

```python
a5 = code_filler(a7, '5', 'Trade and integration')
a5[a5.code == '5'].isnull().sum()
```

```python
a9 = code_filler(a5, '9', 'Urban development')
a9[a9.code == '9'].isnull().sum()
```

```python
a1 = code_filler(a9, '1', 'Economic management')
a1[a1.code == '1'].isnull().sum()
```

```python
a3 = code_filler(a1, '3', 'Rule of law')
a3[a3.code == '3'].isnull().sum()
```
